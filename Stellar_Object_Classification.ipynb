#MODEL BUILDING (BASE MODELS)

Train all ML models using default parameters
(No tuning here)

#Decision Tree

```python
dt_base = DecisionTreeClassifier(random_state=42)
dt_base.fit(X_train, y_train)

dt_pred = dt_base.predict(X_test)
```

#K-Nearest Neighbors

```python
knn_base = KNeighborsClassifier()
knn_base.fit(X_train, y_train)

knn_pred = knn_base.predict(X_test)
```

#Logistic Regression

```python
lr_base = LogisticRegression(max_iter=1000)
lr_base.fit(X_train, y_train)

lr_pred = lr_base.predict(X_test)
```

#Multilayer Perceptron

```python
mlp_base = MLPClassifier(max_iter=300)
mlp_base.fit(X_train, y_train)

mlp_pred = mlp_base.predict(X_test)
```

#Naive Bayes

```python
nb_base = GaussianNB()
nb_base.fit(X_train, y_train)

nb_pred = nb_base.predict(X_test)
```

#Random Forest

```python
rf_base = RandomForestClassifier(random_state=42)
rf_base.fit(X_train, y_train)

rf_pred = rf_base.predict(X_test)
```

#Support Vector Classifier

```python
svc_base = SVC()
svc_base.fit(X_train, y_train)

svc_pred = svc_base.predict(X_test)
```

#soft voting classifier

```python
from sklearn.ensemble import VotingClassifier

voting_clf = VotingClassifier(
    estimators=[
        ('dt', dt_base),
        ('lr', lr_base),
        ('knn', knn_base),
        ('nb', nb_base),
        ('mlp', mlp_base)
    ],
    voting='soft'
)

voting_clf.fit(X_train, y_train)

voting_pred = voting_clf.predict(X_test)
```

#BASE MODEL EVALUATION

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def evaluate(name, y_test, y_pred):
    return {
        "Model": name,
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred, average='weighted'),
        "Recall": recall_score(y_test, y_pred, average='weighted'),
        "F1-Score": f1_score(y_test, y_pred, average='weighted')
    }

base_results = []

base_results.append(evaluate("Decision Tree", y_test, dt_pred))
base_results.append(evaluate("KNN", y_test, knn_pred))
base_results.append(evaluate("Logistic Regression", y_test, lr_pred))
base_results.append(evaluate("MLP", y_test, mlp_pred))
base_results.append(evaluate("Naive Bayes", y_test, nb_pred))
base_results.append(evaluate("Random Forest", y_test, rf_pred))
base_results.append(evaluate("SVM", y_test, svc_pred))
base_results.append(evaluate("Voting Classifier", y_test, voting_pred))


base_results_df = pd.DataFrame(base_results)
base_results_df
```

#HYPERPARAMETER TUNING

```python
from sklearn.model_selection import RandomizedSearchCV

rf_params = {
    'n_estimators': [100, 200],
    'max_depth': [10, 12, 15],
    'min_samples_split': [10, 14],
    'min_samples_leaf': [4, 6],
    'max_features': ['log2']
}

rf_tuned = RandomizedSearchCV(
    RandomForestClassifier(),
    rf_params,
    cv=5,
    n_jobs=-1
)

rf_tuned.fit(X_train, y_train)
```

#TUNED MODEL EVALUATION

```python
rf_best = rf_tuned.best_estimator_
rf_pred_tuned = rf_best.predict(X_test)

evaluate("Random Forest (Tuned)", y_test, rf_pred_tuned)
```

#CREATE A FUNCTION TO PLOT CONFUSION MATRIX

### Confusion Matrix Interpretation

The confusion matrix provides a detailed breakdown of correct and incorrect predictions
made by the classification model. The diagonal elements represent correctly classified
instances, while off-diagonal elements indicate misclassifications.

From the confusion matrix, it is observed that:
- The Random Forest model correctly classifies most of the stellar objects.
- Star class has the highest classification accuracy.
- Most misclassifications occur between galaxies and quasars due to their similar spectral properties.
- The results confirm the robustness and reliability of the trained model.


IMPORT REQUIRED LIBRARIES

```python
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

```python
def plot_confusion(model, X_test, y_test, title):
    plt.figure(figsize=(5,4))
    ConfusionMatrixDisplay.from_estimator(
        model,
        X_test,
        y_test,
        cmap="Blues",
        values_format="d"
    )
    plt.title(title)
    plt.show()
```

Decision Tree

```python
plot_confusion(dt_base, X_test, y_test, "Decision Tree Confusion Matrix")
```

K-Nearest Neighbors

```python
plot_confusion(knn_base, X_test, y_test, "KNN Confusion Matrix")
```

Logistic Regression

```python
plot_confusion(lr_base, X_test, y_test, "Logistic Regression Confusion Matrix")
```

Random Forest

```python
plot_confusion(rf_best, X_test, y_test, "Random Forest Confusion Matrix")
```

Multilayer Perceptron

```python
plot_confusion(mlp_base, X_test, y_test, "MLP Confusion Matrix")
```
